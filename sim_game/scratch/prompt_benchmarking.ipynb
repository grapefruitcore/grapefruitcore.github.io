# Status AI: NovelAI Xialong Prompt Benchmarking & Parsing

This notebook serves as an interactive environment to test prompt templates, verify completion responses, and validate formatting parsers for the Status AI clone.

```python
import sys
import os
# Add parent directory to path so we can import backend packages
sys.path.append(os.path.abspath('..'))

from dotenv import load_dotenv
load_dotenv(dotenv_path='../.env')

from backend import novelai
print("NovelAI Client Base URL:", os.getenv("NOVELAI_BASE_URL"))
```

## 1. Test Timeline Post Parsing

Let's test the timeline post parsing function using sample raw completions from the Xialong model.

```python
sample_timeline_completion = """
kay reposted:
soren
@tsubiclub
* 02:40 AM
trinket by tsubi club summer music 2026 feeling outgoing and party vibe.. does any1 feel this^
----
Ella
@moviesforguyss
* 03:15 PM
Sobbing my eyes out listening to don't mind me by Quadeca and thinking about my ex.
"""

parsed_posts = novelai.parse_generated_posts(sample_timeline_completion)
for post in parsed_posts:
    print(f"Name: {post['name']}")
    print(f"Handle: {post['handle']}")
    print(f"Content: {post['content']}")
    print("-" * 30)
```

## 2. Test DM Vibe Parsing

Let's verify that the dual emotional state parser can extract the character's internal `vibe`, `intensity`, and `text` from the generated block.

```python
sample_dm_completion = """
Vibe: flustered yet amused
Intensity: 65
Message: Oh, stop it! You're just saying that to get a discount on my next gig, aren't you?
"""

parsed_dm = novelai.parse_dm_response(sample_dm_completion)
print("Parsed DM Structure:", parsed_dm)
```

## 3. Test Fallback Parsing

In case the model behaves unpredictably or ignores the format constraints, verify that our parsing logic falls back safely instead of failing.

```python
malformed_dm_completion = "Honestly, I'd love to collaborate on a track with you sometime. Let me know when you're free!"
parsed_fallback = novelai.parse_dm_response(malformed_dm_completion)
print("Parsed Fallback Structure:", parsed_fallback)
```

## 4. Run Live Completion Test (Requires API key)

Try running a live text completion against the NovelAI endpoints using the `xialong-v1` model configuration.

```python
import asyncio

async def test_live_call():
    if not os.getenv("NOVELAI_API_KEY"):
        print("Skip: API key not set.")
        return
    try:
        resp = await novelai.generate_timeline_posts("Underground Music")
        print("Successfully generated timeline posts:", resp)
    except Exception as e:
        print("Error calling API:", e)

# If running inside Jupyter notebook, run the async function
await test_live_call()
```
